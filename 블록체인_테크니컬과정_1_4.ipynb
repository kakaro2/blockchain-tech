## 영지식증명알고리즘

### 개념설명

영지식 증명 알고리즘은 1985년 MIT 연구원이</br>  "The kowledge complexity of interactive proofsystems" 논문에서 처음으로 제안한 개념 </br>
</br> 
영지식증명알고리즘 개념은 증명을 위해서 필요한 직접적인 정보를 제공하지 않고, 진위여부를 입증하는 것임

1. 증명을 함에 있어서 절대 정보를 제공해선 안된다.
2. 이 증명이 충분히 높은 확률로 타당해야한다.
3. 이것을 증명함에 있어서 거짓일 확률이 충분히 낮아야 한다.(0에 가깝다)

눈이 보이지 않는 사람(A)이 있음. 그리고 공 2개를 들고 있고, 2개의 색깔이 같다고 생감함</br>
반면, 이를 보는사람(B)은 공 2개의 색깔이 다르다고 생각함.</br>
A는 B를 믿지 못하는 상황임. 색깔을 말하지 않고 어떻게 증명할까?
</br></br>
A는 공 2개를 숨겨서 처음에는 공을 제시함. 그리고 이후  A는 연속으로 2개 공 중 무작위로 제시함 . 공이 다르다면, B는 올바르게 말할 수 있음.
</br>
그러나 공이 같다면, B는 확률적으로 틀릴 수 밖에 없음. A가 선택한 공을 B가 못보기 때문임</br>
이러한 확률은 5번 시도(2^5)에서 1/32로 B는 A를 속일 수 있음. 10(2^10)번 시도에는 1/1024확률로 B가 A에게 공이 다르다고 속일 수 있음.

### 블록체인 적용 내용



영지식증명알고리즘은 상호작용형과 비상호작용형이 있음.
- 상호작용형은 입증자 A와 상대방 B 사이에 영지식 정보 교환을 통해 이뤄짐
- 반면, 비상호작용형은 A가 B에게 일방적으로 전달하여 증명할 수 있음. 이는 작업량이 적음. 그래서 블록체인 암호학 기술로 사용됨.

개인정보보호 , 확장성

## 블록체인 확장성

### 트릴레마 관점에서의 블록체인 한계
비트코인의 '확장성(Scalability, 낮은 TPS)에 한계가 제기됨</br> -> 이를 해소하기 위한 연구 진행

### Segwit
Segwit은 디지털 전자 서명 부분을 별도로 분리해 블록에 담을 수 있는 거래 크기를 증가시킨 방식

- 거래 가변성 문제 해결(부인 방지 효과)</br>
- 전자 서명 데이터가 전체 데이터의 75%(?) -> 최대 4MB로 키우는 효과를 가져옴

### Off-Chain solution
스테이트 채널 블록체인은 기존 블록체인 확장성을 해결하기 위해 고안한 오프-블록체인 솔루션

image.png

### Lightening Network
비트코인 기반의 또 다른 레이어를 만들어 소위 그룹간에 비트코인 결제를 가능케 하는 네트워크

소액 결제를 빠르게 처리하는 오프체인 솔루션(2015년 처음으로 제안됨)</br>
멀티 시그널: 거래 안정성을 위해 거래 참여자가 양방향 으로 동의해야 함

image.png

### Raiden Network
별도의 체인을 만들어 모두의 합의를 거치지 않아도 되는 거래 기술 (이더리움 전용)

잔액증명
- 거래 당사자간의 거래가 잔액을 초과하지 않게 기록

### Plasma
오프체인 방식 (헤더의 묶음별로 다차원적으로 형성하는 블록체인)

루트체인: 이더리움의 메인 체인</br>
차일드 체인은 페런트 체인에게 거래 헤더를 주기적으로 제공함</br> 
- 라이트닝, 라이덴 등 보다 더 높은 무결성 제공 


지원자 역할

### 샤딩
샤딩으로 블록체인의 확장성 문제를 해결할 수 있음

샤딩은 데이터베이스에 사용된 개념으로, 대용량 처리를 위해 데이터를 분할하는 개념</br>
샤딩은 온체인 솔루션인 반면, 스테이트 채널은 오프체인 솔루션

## 정리

스마트 컨트랙트
- 계약 방식을 자연어에서 컴퓨터 언어로 대체하기 위해 등장한 기술
- 무결성 부분의 취약점문제가 제기됨. 이런 문제를 블록체인으로 해결
- 스마트 컨트랙트 등장은 블록체인을 화폐에서 계약으로 확장하게 하는 계기가 됨

영지식증명알고리즘
- 정보를 직접적으로 드러내지 않고 증명하는 알고리즘
- 비밀번호 저장에 사용돼 왔음
- 블록체인에서 개인정보보호로서 활용될 것으로 전망

블록체인 확장성
- 세그윗은 블록 용량을 줄여 확장성을 개선함
- 오프체인솔루션은 블록체인 밖에서 이뤄지는 거래 방식. 확장성 향샹
- 샤딩은 합의 과정을 쪼개어 성능을 향상한 방식
