## 정리


### 블록체인과 행정학
- 합의 알고리즘은 의사결정 방식에 있어서 행정학의 철학을 담고있음
- 공정한 의사결정이 가능하나, 의사 결정 방식이 느림
- 탈중앙형 합의 알고리즘은 중앙형 클라우드와 비교됨. 서로 간의 장단점이 있음

### 블록체인과 온라인 투표
- 투표 자체에 합의 작동 방식이 내포돼 있음. 이에 따라, 온라인 투표 자체에 합의 알고리즘 모습이 보임.
- 온라인 투표 BM은 투표결과에서 발생할 수 있는 조작 위험
- 아고라 보팅, 따복 등 블록체인 성공 사례가 이미 많이 존재함

### 블록체인 비즈니스 가치
- 블록체인의 비즈니스 가치는 합의 알고리즘(무결성)에서 비롯됨
- <투명성 + 무결성>은 새로운 신뢰성이라는 비즈니스 가치를 제공하고, 이에 따른 신규 BM 발굴이 가능
- 중개 기관 대체, 이력 추적 등에서 BM을 확인할 수 있음
