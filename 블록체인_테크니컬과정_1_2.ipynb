## 합의알고리즘 구분 방법

### 1.FLP 불가능성 이론

#### 기본설명

FLP 불가능성은 "Impossiblilty of Distributed Consensus with One Faulty Process(1985)" 에서 처음으로 3가지 요인을 모두 만족할 수 없음을 입증했음.

FLP의 3요인
- Safety -> Guaranteed Termination
 - 완전성(Finalty)으로 표현하기도 함
 - 해당 요구 사항은 시스템 노드 사이에서 공유되는 가치는 모두 일치하고 유효함
 - 선합의 후해결
- Liveness -> Not Guaranteed Termination
 - 생동성을 뜻하는 용어로 시스템의 원활성을 위해 필요한 요건
 - 블록체인에서는 끊임 없이 블록을 원할하게 생성해야 하는 조건
- Fault Tolerance
 - 노드 장애에도 시스템은 정상적으로 돌아가야 함을 의미 (비잔틴 장군 문제 해결)

image.png

예시) 비트코인
- 컴퓨팅 파워(PoW)로 Fault Tolerance 해결.
- 그러나 Liveness 우선. 블록 생성 우선
- Safety 보장 X (i.e 포크 발생 가능) -> 6 Confirmation

</br>예시) BFT 계열
- 일정 참여자 조건에서 2/3 이상 동의를 얻지 못하면 블록 생성 X. Fault Tolerance 해결.
- Safety 보장. Liveness 보장 X

합의 알고리즘 개론
- 이중지불 대응을 위해 등장한 기술
- 합의 알고리즘은 탈중앙 체계에서 공정하게 운영과 관리를 위해 기술로 정의할 수 있음
- 합의 알고리즘은 블록체인의 전부라고 할 정도로 비중이 큰 기술

합의 알고리즘 구분 방법
- FLP 불가능 이론은 합의 알고리즘 유형을 "생동성 우선"과 "실시간성 우선"으로 나누게 함
- 실시간성 우선은 블록을 주기적으로 생성하는데, 넷스플리트로 인한 포크 발생 유발
- 이는 취약점 발생 유발, 이중지불공격으로 인한 피해 유발
